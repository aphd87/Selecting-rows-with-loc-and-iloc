```python
import pandas as pd
transactions = pd.read_excel("grocery_database.xlsx", sheet_name = "transactions")
```

```python
#transactions.loc[row_labels, column_label]
#transactions.iloc[row_indexes, column_indexes]

#ILOC
#Method for specifiying conditions on rows and columns by using indexes
#I - Index
transactions.loc[0] #We are returned the entire row

```

```python
#Sequence of rows
transactions.loc[0:4] #We are returned 0-3 rows
```

```python
#Selection of rows 0, 30, 51
transactions.iloc[[0,30,51]] 
```

```python
#Selection of rows 0:4
#Selection of columns 0, 3, -1
transactions.iloc[0:4,[0,3,-1]]
```

```python
#All rows, select columns
#Lone colon allows for this
transactions.iloc[:,[0,3,-1]]
```

```python
#LOC
#Uses names of columns
#Method for specifiying conditions on rows and columns that use names
#Conditional Logic
```

```python
transactions.loc[0]
#Index is 0 in data set, so we get the same result.
#The label is also named 0 here.
```

```python
transactions.set_index("customer_id", inplace=True)
```

```python
transactions.loc[642]
#All rows for customer 642
#Want index to have unique values, not duplicates here
```

```python
transactions.reset_index(inplace=True) #Reset index
```

```python
list(transactions) #List of column names
#This is helpful before conducting LOC
```

```python
transactions.loc[0:10, "customer_id"]
#We want rows 0:10, want customer_id column to be returned
```

```python
transactions.loc[0:10, ["customer_id", "product_area_id", "sales_cost"]]
```

```python
#Manipulate order here
transactions.loc[0:10, [ "product_area_id", "sales_cost","customer_id",]]
```

```python
#Keep rows when something is true 
transactions["customer_id"] == 642 #Where the condition is true
#A signle equal sign will assign 642 to all rows - we don't want this

transactions.loc[transactions["customer_id"] == 642]
#Only return rows where this condition is true - filter rows on certain condition


```

```python
#Specify specific columns here with the condition still intact
transactions.loc[transactions["customer_id"] == 642, ["customer_id", "sales_cost"]]
```

```python
#Specify two conditions that must occur together (AND)
transactions.loc[(transactions["customer_id"] == 642) & (transactions ["num_items"] > 5)]
#Pandas needs us to enclose each condition with parentheses
```

```python
#We need one of the conditions to be true (OR) - all rows where C ID is 642 OR greater than 5
transactions.loc[(transactions["customer_id"] == 642) | (transactions["num_items"] > 5)]
```

```python
#IS IN
#Provide list of values as part of condition
#If we want 642, but also 700, we can create multiple statements here
transactions.loc[(transactions["customer_id"] == 642) | (transactions["num_items"] > 5)]
```

```python
#We are able to get multiple items from one column
transactions.loc[transactions["customer_id"].isin([642,700])]
```

```python
#If we wanted all rows where customer rows WERE NOT 642, 700
#We add a tilda next to transactions
transactions.loc[~transactions["customer_id"].isin([642,700])]
```
